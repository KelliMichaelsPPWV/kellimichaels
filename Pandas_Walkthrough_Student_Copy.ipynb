##### User Defined Function

<p>If what you want to do to a column that can't be represented by simple mathematical operations, you can write your own $\textit{user defined function}$ with the full customizability available in Python and any external Python packages, then map it directly onto a column. Let's add some ages to our customer dataframe, and then classify them into our custom defined grouping scheme:</p>

```python
#remember our people are in a dataset called data, don't use the original dictionary
#my user defined function is going to calculate what year people weere born, and then I'll use that to add my new column

def calculate_birth_year(age):
    current_year = pd.Timestamp.now().year
    return current_year - age

# Apply the function and create a new column 'birth_year'
data['birth_year'] = data['ages'].apply(calculate_birth_year)

data
```

<p>As a last example I'll show here how you would use a lambda function to create a UDF that depends on $\textit{more than one}$ column:</p>

<li>UDF = User Defined Function</li>

```python
#add another silly column to data, I am going to add their age and number of letters in their name together
#depends on name string and age int

def calculate_total(row):
    name_length = len(row['names'])  
    age = row['ages'] 
    return age + name_length  

#use the apply functiona and axis = 1 indicates row-wise operations
data['total'] = data.apply(calculate_total, axis=1)
data
```

### In-Class Exercise #3 - Create Your Own UDF <br>

```python
"""
    Name  BA OBP AllStar
    --------------------
    Name .233 .360 Yes
    Name .150 .288 No
"""

##column names are BA anf OBP

def is_all_star(row):
    if row['BA'] > .280 or row["OBP"] > .360:
        return "yes"
    else :
        return "no"

bos17["All-Star?"]=bos17.apply(is_all_star, axis = 1)    

bos17

bos18["All-Star?"]=bos18.apply(is_all_star, axis = 1)    

bos18

```

### Homework Excersise #1 - Find the Total Number of Runs and RBIs for the Red Sox <br>
<p>Get total number of home runs and rbi's</p>

```python
# step 1: Add a new column with the key 'Team' and all column values should be 'BOS'
# step 2: Group by the 'Team' column and get total home runs and rbi's
# Produce data for both 2017 and 2018 (ie print both seperated by a newline character \n)

# Step 1: Add a new column 'Team' with the value 'BOS' to both DataFrames
bos17['Team'] = 'BOS'
bos18['Team'] = 'BOS'

# Step 2: Group by 'Team' and sum the 'HR' and 'RBI' columns for both years
bos17_grouped = bos17.groupby('Team')[['HR', 'RBI']].sum()
bos18_grouped = bos18.groupby('Team')[['HR', 'RBI']].sum()

print(bos17_grouped)
print("\n")
print(bos18_grouped)
```
