```python
import requests

##I TOOK THIS FROM THE INTERNET I JUST WANTED A NICE EXAMPLE WORK-FLOW, THIS BOX IS NOT MY WORK

# The URL to the API for 2008 Formula 1 driver standings after round 5
url = 'http://ergast.com/api/f1/2008/5/driverStandings.json'

# Send a GET request to the API
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the response as JSON
    data = response.json()
    
    # Navigate to the part of the JSON where the standings are located
    standings_list = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
    
    # Loop through the standings and extract specific information
    for driver in standings_list:
        driver_name = driver['Driver']['givenName'] + " " + driver['Driver']['familyName']
        points = driver['points']
        position = driver['position']
        nationality = driver['Driver']['nationality']
        
        # Print out the data
        print(f"Position: {position}, Name: {driver_name}, Points: {points}, Nationality: {nationality}")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
##I want to make 20 pokemon by using the pokeAPI
##first let's get a handle on the organization on the API structure and what data we need to create for each guy
##we want a dictionary that looks like:
###{ 'Fire'
##        {'guy': {'abilities', 'weight'..}}
##        {'guy2': {'abilities', 'weight'..}}}
##  'Flying':
##      {'guy3': {'abilities', 'weight'..}}
##      {'guy4': {'abilities', 'weight'..}}}
## ......}

##I used this block for printing and poking around to get all the structure
##poke_url = f"https://pokeapi.co/api/v2/pokemon/ivysaur"
##responded = requests.get(poke_url)
##poke_data = responded.json()
##print(json.dumps(poke_data))

#Atual coding a functions start here
#function for reading in data from API
def get_pokemon_data(pokemon_name):
    poke_url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}/"
    responded = requests.get(poke_url)
    poke_data = responded.json()
    ##name and weight are accessible right away
    name = poke_data['name']
    weight= poke_data['weight']
    ## abilities and types need to be read from the subcatgies of abilities, look inside for name of ability
    ##look inside of type for the name of the type
    abilities = [ability['ability']['name'] for ability in poke_data['abilities']]
    types = [poke_type['type']['name'] for poke_type in poke_data['types']]
    return {
            'name': name,
            'weight': weight,
            'abilities': abilities,
            'types': types
        }
##initialize poketype dictionary
type_dict={}

#grab the first twenty, starting at one not zero
#and then order in the type dictiory BY TYPE, so not by name/number like original
for poke_id in range(1,21):
    poke_data = get_pokemon_data(poke_id)
    #once we have sucessfully retrieved data it is time to sort
    if poke_data:
        for poke_type in poke_data['types']:
            # If this type is not already a key in type_dict, create it
            if poke_type not in type_dict:
                type_dict[poke_type] = {}
            
            # Add the Pokémon to the dictionary under its type
            type_dict[poke_type][poke_data['name']] = {
                'abilities': poke_data['abilities'],
                'weight': poke_data['weight']
            }  

print(json.dumps(type_dict, indent = 4))
```
